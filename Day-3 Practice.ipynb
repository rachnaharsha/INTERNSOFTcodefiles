```python
#prime number

def pn(num):
    for i in range(2,num):
        if num%i == 0:
            print("Composite number")
            break
    else:
        print("Prime number")
            
            
```

```python
pn(277)
```

```python
import math 

def pn2 (num):
    if (num%2)==0 and num>2:
        return False
    for i in range(3, int(math.sqrt(num))+1,2 ):
        if num%i==0:
            return False
    return True
        
    
```

Iterators and Generators

```python
def gen(num):
    for i in range(0,num):
        yield (i**3)
    
```

```python
for x in gen(12):
    print (x)
```

```python
#fibonacci series

def fibo(num):
    a=1
    b=1
    for i in range (num):
        yield (a)
        a,b=b,a+b
```

```python
for x in fibo(9):
    print(x)
```

```python
#normal fibonacci function 

def fibon (num):
    a=1
    b=1
    for i in range (num):
        print(a)
        a,b=b,a+b
    
```

```python
fibon(10)
```

map()

```python
def fahrenheit(T):
    return ((float(9)/5)*T + 32)
def celsius(T):
    return (float(5)/9)*(T-32)
    
temp = [0, 22.5, 40,100]
```

```python
F_temps = list(map(fahrenheit, temp))
F_temps
```

```python
list(map(celsius,F_temps))
```

```python
list(map(lambda x: (5/9)*(x-32), F_temps))
```

filter()

```python
#Prime number 
import math
def prime_check(num):
    if (num%2==0) and (num>2):
        return False
    for i in range(3,int(math.sqrt(num))+1,2):
        if (num% i)==0:
            return False
    return True
```

```python
lst = range(20)
list (filter(prime_check,lst))
```

```python
print(list(map(prime_check,lst)))
list(lst)
```

```python
#to print odd numbers
list (filter(lambda x: x%2==1,lst))
```
