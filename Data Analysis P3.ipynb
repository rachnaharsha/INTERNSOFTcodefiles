# Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

plt.plot(np.random.normal(size=10),np.random.normal(size=10),'^')
print(np.random.normal(size=10))
```

```python
%matplotlib inline
normals=pd.Series(np.random.normal(size=10))
print(normals)
normals.plot(grid=True)
```

```python
normals.cumsum().plot(grid=True)
```

```python
var = pd.DataFrame({'normal': np.random.normal(size=5),'gamma': np.random.gamma(1, size=5),'poisson': np.random.poisson(size=5)})
print(var)
var.cumsum(0).plot(grid=True)
```

```python
var.cumsum(0).plot(subplots=True, grid=True)
```

```python
var.cumsum(0).plot(secondary_y='poisson')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 4))
for i,v in enumerate(['normal','gamma','poisson']):
    var[v].cumsum(0).plot(ax=axes[i], title=v)
axes[0].set_ylabel('cumulative sum')
```

```python
fig.savefig('figure.png')
```

```python
plt.plot(s,np.sin(s),'r')
plt.plot(s,np.cos(s-1),'y')
plt.plot(s,np.sin(s+1),color='b')
plt.plot(s,np.sin(s-2),'0.1')
```

```python
plt.plot(s,np.sin(s),linestyle=':')
plt.plot(s,np.cos(s-1),linestyle='-.')
plt.plot(s,np.sin(s+1),color='b')
```

## Simple scatter plots

```python
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(1, 10, 20)
y = np.log(x)

plt.plot(x, y, 'o:', color='violet')
```

```python
plt.plot(x,y,'-or')
```

```python
plt.scatter(x,y+2)
```

```python
rng = np.random.RandomState(5)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 100 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=1,
            cmap='Blues')
plt.colorbar()  # show color scale
```

### Errorbars

```python
a=np.linspace(0,100,20)
dy=0.1
y = np.sin(x) + dy * np.random.randn(100)
plt.errorbar(x, y, yerr=dy, fmt='.');
```

### Histograms

```python
plt.style.use('seaborn-white')
data=np.random.randn(100)
print(data)
```

```python
plt.hist(data)
```

```python
plt.hist(data,bins=30,alpha=0.6,histtype='stepfilled')
```

```python
x=np.random.normal(0,0.8,1000)
y=np.random.normal(-2,1,1000)
kwargs = dict(histtype='stepfilled', alpha=0.7, bins=50)
plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```python
mean=[5,3]
cov=[[3,1],[2,0.5]]
x,y=np.random.multivariate_normal(mean,cov,10000)
```

```python
plt.hist2d(x,y,bins=20,cmap='Blues')
b=plt.colorbar()
```
